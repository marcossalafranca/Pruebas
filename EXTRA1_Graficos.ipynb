## Eje cartesiano de coordenadas

Pensemos en la recta real. Los enteros de –3 a 3 están marcados en la línea, pero entre cualquiera de estos dos números (por ejemplo, 1 y 2) se encuentran todos los números posibles: 1.0, 1.1, 1.2, 1.3 ... y no sólo al nivel de décimas, sino todos los números con dos decimales, tres decimales,..., infinitos decimales.

<img src="Images/recta_real.png" style="width: 300px;"/>

La recta real hace que ciertas propiedades sean visualmente intuitivas:
* Por ejemplo, todos los números en el lado derecho de 0 son positivos, y los del lado izquierdo son negativos.  
* Cuando un número a se encuentra en el lado derecho de otro número b, a siempre es mayor que b y b siempre es menor que a.  
* Las flechas en los extremos de la línea numérica indican que la línea se extiende infinitamente, y cualquier punto en esta línea corresponde a algún número real, por grande que sea.  
* Un solo número es suficiente para describir un punto en la recta numérica.  

Ahora consideremos los ejes cartesianos en $\mathbb{R}$<sup>2</sup> . Las líneas numéricas se intersecan en ángulos rectos entre sí y se cruzan en el punto 0 de cada línea. Esto forma un plano de coordenadas cartesianas, o un plano x-y, con la línea numérica horizontal llamada eje x y la línea vertical llamada eje y.

<img src="Images/Plano_Cartesiano_image001.jpg" style="width: 300px;"/> 

Al igual que con la recta real, podemos tener infinitos puntos en el plano. Describimos un punto con un par de números en lugar de un número. Por ejemplo, describimos el punto P con dos números, x e y, escritos como (x, y) y denominados las *coordenadas del punto*.  

<img src="Images/plano_cartesiano2.gif" style="width: 300px;"/>  

Como podemos ver en la imagen, x es la distancia del punto desde el origen a lo largo del eje x, e y es la distancia a lo largo del eje y. El punto donde se intersecan los dos ejes se denomina origen y tiene las coordenadas (0, 0).
El plano de coordenadas cartesiano nos permite visualizar la relación entre dos conjuntos de números. Aquí, uso el término conjunto para significar una colección de números. No importa lo que representen los dos conjuntos de números (temperatura, puntuaciones, edades, etc.) todo lo que necesitamos son los números en sí. Luego, puede trazarlos, ya sea a mano o computacionalmente.  


## Creando gráficos con MatplotLib 

Usaremos matplotlib para hacer gráficos con Python. Matplotlib es un paquete de Python útil para trazar números y hacer gráficos.

<div class="alert alert-success">
    <b>Comenzaremos con un gráfico simple con solo tres puntos: (1, 2), (2, 4) y (3, 6)</b><br><br>
</div> 

Para crear este gráfico, primero haremos dos listas de números: 
- uno que almacena los valores de las coordenadas $x$ de estos puntos.
- otra que almacena las coordenadas $y$.

```python
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
```

Lo graficamos

```python
import matplotlib.pyplot as plt
plt.plot(x_numbers, y_numbers)
plt.show()
```

Démonos cuenta que en lugar de comenzar desde el origen (0, 0), el eje x comienza desde el número 1 y el eje y comienza desde el número 2. Estos son los números más bajos de cada una de las dos listas.  

<div class="alert alert-success">
    <b>Marcar los puntos que dimos para el trazado</b>
</div> 

```python
plt.plot(x_numbers, y_numbers, marker='o')
plt.show()
```

Podemos elegir entre varias opciones de marcadores. Veamos algunos de ellos:

```python
plt.plot(x_numbers, y_numbers, marker='*')
plt.show()
```

```python
plt.plot(x_numbers, y_numbers, marker='x')
plt.show()
```

```python
plt.plot(x_numbers, y_numbers, marker='+')
plt.show()
```

Como hemos visto, por defecto plot incluye una línea que conecta los puntos. Sin embargo, sepuede hacer un gráfico que marque solo los puntos que especificamos, sin ninguna línea que los conecte, omitiendo la palabra "marker":

```python
plt.plot(x_numbers, y_numbers, 'o')
plt.show()
```

```python
temperaturas = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3] 
plt.plot(temperaturas, marker='o')
```

Almacenamos las temperaturas medias en una lista, temperaturas. Luego, llamamos a la función plot() pasando solo esta lista y el marcador. Cuando usas plot() en una sola lista, esos números se trazan automáticamente en el eje y. 
    
Los valores correspondientes en el eje x se completan como las posiciones de cada valor en la lista. Es decir, el primer valor de temperatura, 53.9, obtiene un valor de eje x correspondiente de 0 porque está en la posición 0 de la lista. Como resultado, los números trazados en el eje x son los números enteros de 0 a 12, que podemos considerar que corresponden a los 13 años para los cuales tenemos datos de temperatura.

Si nos fijamos ahora en el gráfico, vemos cómo muestra si la temperatura promedio ha subido y bajado con respecto al año anterior. Si nos fijamos en los números que pintamos, realmente no son muy lejanos el uno del otro. Sin embargo, el gráfico hace que las variaciones parezcan bastante dramáticas. Entonces, ¿qué está pasando?

La razón es que matplotlib elige el rango del eje y de modo que sea suficiente para encerrar los datos suministrados para el trazado.  

En este gráfico, el eje y comienza en 53.0 y su valor más alto es 57.5.Esto hace que incluso las pequeñas diferencias parezcan dramáticas porque el rango del eje y es muy pequeño. 

Graficar la temperatura sin mostrar los años correspondientes es una forma rápida y fácil de visualizar las variaciones entre los años. Sin embargo, si queremos presentar este gráfico a alguien, sería ideal mostrarlas:

```python
temperaturas = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3] 
anyos= range(2000, 2013)
plt.plot(anyos, temperaturas, marker='o')
```

<div class="alert alert-success">
Seguimos con la ciudad de Nueva York. Veamos ahora cómo ha variado la temperatura mensual promedio a lo largo de los años. Para los años 2000, 2006 y 2012 tracemos la temperatura promedio de los 12 meses.
Primero, necesitamos crear tres listas para almacenar la temperatura (en Fahrenheit). Cada lista constará de 12 números correspondientes a la temperatura promedio de enero a diciembre de cada año:
</div> 

```python
temperaturas_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1] 
temperaturas_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6] 
temperaturas_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]
```

Podríamos trazar los tres conjuntos de datos en tres gráficos diferentes, pero eso no haría que sea muy fácil ver cómo cada año se compara con los demás.

```python
meses = range(1, 13)
plt.plot(meses, temperaturas_2000, meses, temperaturas_2006, meses, temperaturas_2012)
```

Luego, podemos ingresar múltiples pares de listas en la función `plot()`. Con cada lista separada por una coma, la función plot () trazará automáticamente una línea diferente para cada par.  

Python elige automáticamente un color diferente para cada línea para indicar que las líneas se han trazado desde diferentes conjuntos de datos.  

En lugar de llamar a la función de trazado con los tres pares a la vez, también podríamos llamar a la función de trazado tres veces separadas, una para cada par.  

```python
plt.plot(meses, temperaturas_2000) 
plt.plot(meses, temperaturas_2006) 
plt.plot(meses, temperaturas_2012) 
```

Sin embargo, tenemos un problema porque no tenemos ninguna pista sobre qué color corresponde a qué año. Para solucionar esto, podemos usar la función legend(), que nos permite agregar una leyenda al gráfico. Una leyenda es un pequeño cuadro de visualización que identifica lo que significan las diferentes partes de la gráfica.  

En nuestro caso, usaremos una leyenda para indicar qué año representa cada línea de color.

```python
meses = range(1, 13)
plt.plot(meses, temperaturas_2000, meses, temperaturas_2006, meses, temperaturas_2012)
plt.legend([2000, 2006, 2012])
plt.show()
```

Hasta ahora, hemos permitido que Python determine automáticamente los números en ambos ejes en función de los datos suministrados a la función plot(). Esto puede estar bien para la mayoría de los casos, pero a veces este rango automático no es la forma más clara de presentar los datos, como vimos en el gráfico donde representamos la temperatura promedio anual de la Ciudad de Nueva York. ### Ejemplo: La ley de gravitación universal de Newton.

Queremos ver qué pasa con la fuerza a medida que aumenta la distancia entre los dos cuerpos.

<img src="Images/ley-de-la-gravitacion-universal.jpg" style="width: 300px;"/>  

<div class="alert alert-success">
            Tomemos las masas de dos cuerpos: la masa del primer cuerpo $(m_1)$ es 0.5 kg, y la masa del segundo cuerpo $(m_2)$ es 1.5 kg. El valor de la constante gravitacional es 6.674 × 10−11 N m2 kg<sup>−2</sup>.<br><br>
            
    Vamos a calcular la fuerza gravitacional entre estos dos cuerpos a 19 distancias diferentes: 100 m, 150 m, 200 m, 250 m, 300 m, y así sucesivamente hasta 1000 m.
    </div> 

```python
import matplotlib.pyplot as plt

# Definimos el gráfico
def grafico(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distancia en m')
    plt.ylabel('Fuerza gravitacional en Newtons')
    plt.title('Fuerza gravitacional vs distancia')
    plt.show()

# Generamos el rango de distancias
def generador_F_r():
    r = range(100, 1001, 50)
    # Creamos la lista vacia F para almacenar los valores calculados de F 
    F=[]
    # Declaramos la gravedad
    G = 6.674*(10**-11)
    # Ahora los dos valores de las masas
    m1 = 0.5
    m2 = 1.5
    # Calculamos las fuerzas y las añadimos a la lista F 
    for dist in r:
        fuerza = G*(m1*m2)/(dist**2)
        F.append(fuerza)
    # Llamamos a la función grafico
    grafico(r, F)

generador_F_r()
```

A medida que aumenta la distancia `r`, la fuerza gravitacional disminuye.
    
Con este tipo de relación, decimos que la fuerza gravitatoria es inversamente proporcional a la distancia entre los dos cuerpos. Además, tengamos en cuenta que cuando el valor de una de las dos variables cambia, la otra variable no necesariamente cambiará en la misma proporción. Es lo que se denomina una relación no lineal.  
Como resultado, terminamos con una línea curva en el gráfico en lugar de una recta.
